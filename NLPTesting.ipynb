Import Step:

```python
import pandas as pd
import numpy as np

#When loading datasets for nlp it is not necerssary to use dataframes as typical ML so far
data = open(r'Datasets\AI Generated User stories\universitymanagement.csv').read()
```

spaCy example:

```python
import spacy

# Load English tokenizer, tagger, parser and NER
nlp = spacy.load("en_core_web_sm")

# Process whole documents
doc = nlp(data)

# Analyze syntax
print("Noun phrases:", [chunk.text for chunk in doc.noun_chunks])
print("Verbs:", [token.lemma_ for token in doc if token.pos_ == "VERB"])

# Find named entities, phrases and concepts
for entity in doc.ents:
    print(entity.text, entity.label_)


```

```python
for token in doc:
    print(token.text)
```

```python
 
# Iterate over the tokens
for token in doc:
    # Print the token and its part-of-speech tag
    print(token.text, "-->", token.pos_)
```

```python
spacy.explain("PUNCT") 
```

```python

# dependency parsing
for token in doc:
    print(token.text, "-->", token.dep_)
```

```python
spacy.explain("nsubj"), spacy.explain("ROOT"), spacy.explain("aux"), spacy.explain("advcl"), spacy.explain("dobj")
```

```python
from spacy.matcher import Matcher
matcher = Matcher(nlp.vocab)

pattern = [{'TEXT': 'user', 'POS': 'NOUN'}]

# Initialize the matcher with the shared vocab
matcher = Matcher(nlp.vocab)
matcher.add('rule_2', None, pattern)
```

```python
import spacy
from collections import Counter
nlp = spacy.load('en_core_web_sm')
# all tokens that arent stop words or punctuations
words = [token.text
         for token in doc
         if not token.is_stop and not token.is_punct]

# noun tokens that arent stop words or punctuations
nouns = [token.text
         for token in doc
         if (not token.is_stop and
             not token.is_punct and
             token.pos_ == "NOUN")]

# five most common tokens
word_freq = Counter(words)
common_words = word_freq.most_common(5)

# five most common noun tokens
noun_freq = Counter(nouns)
common_nouns = noun_freq.most_common(5)

print (common_nouns)
print (common_words)
```

Used to count types of words, punctuations, etc.

Possible to use this method for feature generation

```python
import spacy.attrs

# Returns integers that map to parts of speech
counts_dict = doc.count_by(spacy.attrs.IDS['POS'])

# Print the human readable part of speech tags
for pos, count in counts_dict.items():
    human_readable_tag = doc.vocab[pos].text
    print(human_readable_tag, count)
```

```python
# Add match ID "HelloWorld" with no callback and one pattern
pattern = [{"POS": "NOUN"}, {"IS_PUNCT": True}, {"TEXT": "I"}]
matcher.add("HelloWorld", [pattern])

matches = matcher(doc)
for match_id, start, end in matches:
    string_id = nlp.vocab.strings[match_id]  # Get string representation
    span = doc[start:end]  # The matched span
    print(match_id, string_id, start, end, span.text)
```
